```python
X = df.drop(['Is Fraudulent'], axis=1)

y = df['Is Fraudulent']
```

```python
from imblearn.over_sampling import RandomOverSampler
ro=RandomOverSampler()
x_data,y_data=ro.fit_resample(X,y)
```

```python
from collections import Counter
print("Actual Data:",Counter(y))
```

```python
print("Artificial Data:",Counter(y_data))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size=0.2,random_state=10)
```

```python
import xgboost as xgb
```

```python
rfc = xgb.XGBClassifier()
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = rfc.predict(x_test )
accuracy_score(y_pred,y_test) 
```

```python
import sklearn.metrics
```

```python
print(sklearn.metrics.classification_report(y_test, y_pred))
```

```python
y_pred = rfc.predict(x_test )
y_true=y_test


from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_true,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```

```python
import pickle

pickle.dump(rfc,open('fraud_xg.pkl','wb'))
xgboost = pickle.load(open('fraud_xg.pkl','rb'))
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier


estimators = [
    ('rf', RandomForestClassifier()),  
    ('dt', DecisionTreeClassifier())   
]


stacking_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())


stacking_model.fit(x_train, y_train)
```

```python
stacking_model.score(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = stacking_model.predict(x_test )
accuracy_score(y_pred,y_test) 
```

```python
print(sklearn.metrics.classification_report(y_test, y_pred))
```

```python
y_pred = stacking_model.predict(x_test )
y_true=y_test


from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_true,y_pred)
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax=plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,linewidths=0.5,linecolor="red",fmt=".0f",ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_true")
plt.show()
```
